```python
import pandas as pd
```

```python
raw_data = pd.read_csv("Absenteeism_data.csv")
```

```python
raw_data
```

```python
df = raw_data.copy()
```

```python
df.info()
```

```python
df = df.drop(["ID"], axis=1)
df
```

### Create dummies for the "Reason for Absence" column

```python
reason_columns = pd.get_dummies(df["Reason for Absence"], drop_first=True)
## We drop the first one to avoid multicollinearity
reason_columns
```

## Group the Reasons for Absence

```python
df.columns.values
```

```python
reason_columns.columns.values
### If we add 28 new columns the dataset will be too large!
```

```python
df = df.drop(["Reason for Absence"], axis=1)
df
```

```python
## We can group the whole 28 variables into groups that are similar
reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:].max(axis=1)
reason_type_1
```

## Concatenate Column values

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', "Reason 1", "Reason 2", "Reason 3", "Reason 4"]
```

```python
df.columns= column_names
df
```

## Reordering the columns

```python
column_names_reordered = ["Reason 1", "Reason 2", "Reason 3", "Reason 4", 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_names_reordered]
df
```

## Create a Checkpoint

```python
## This is for save the current status of the data for further changes
## It's important because we finish our work with "reason for absence" column
df_reason_mod = df.copy()
df_reason_mod
```

## Preprocessing 'Date' column

```python
type(df_reason_mod["Date"][0])
```

```python
## we have to cast the date to timestamp
df_reason_mod["Date"] = pd.to_datetime(df_reason_mod["Date"], format="%d/%m/%Y")
df_reason_mod
```

```python
type(df_reason_mod["Date"][0])
```

```python
df_reason_mod.info()
```

## Extracting the Month value

```python
df_reason_mod["Date"][0]
```

```python
df_reason_mod["Date"][0].month
```

```python
list_months = []
list_months
```

```python
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod["Date"][i].month)
```
